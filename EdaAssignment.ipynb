```python
import pandas  as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df.head()
```

What is the distribution of age among heart failure patients in the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution of Age Among Heart Failure Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

What is the percentage of male and female patients in the dataset?

```python
male = (len(df[df['sex'] == 1])/len(df))*100
male
```

```python
female = (len(df[df['sex'] == 0])/len(df))*100
female
```

```python
plt.figure(figsize=(10, 6))
death_by_age_group = df.groupby('age')['DEATH_EVENT'].mean().reset_index()

sns.barplot(x='age', y='DEATH_EVENT', data=death_by_age_group)
plt.title('Death Count Variation Among Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Mean death Count')
plt.show()
```

How does the platelet count vary among different age groups

```python
plt.figure(figsize=(10, 6))
platelet_by_age_group = df.groupby('age')['platelets'].mean().reset_index()

sns.barplot(x='age', y='platelets', data=platelet_by_age_group)
plt.title('Platelet Count Variation Among Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Mean Platelet Count')
plt.show()
```

Is there a correlation between creatinine and sodium levels in the blood

```python

correlation = df[['serum_creatinine', 'serum_sodium']].corr().iloc[0, 1]
print(f'Correlation between Serum Creatinine and Sodium: {correlation:.2f}')

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df)
plt.title('Correlation Between Serum Creatinine and Sodium Levels')
plt.xlabel('Serum Creatinine')
plt.ylabel('Serum Sodium')
plt.show()


# There is No correlation 
```

How does the prevalence of high blood pressure differ between male and female patients

```python

bp_by_gender = df.groupby('sex')['high_blood_pressure'].mean().reset_index()
bp_by_gender['High Blood Pressure (%)'] = bp_by_gender['high_blood_pressure'] * 100


plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='High Blood Pressure (%)', data=bp_by_gender)
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xlabel('Sex (0: Female, 1: Male)')
plt.ylabel('High Blood Pressure (%)')
plt.show()
```

What is the relationship between smoking habits and the occurrence of heart failure

```python

smoking_death_rate = df.groupby('smoking')['DEATH_EVENT'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='smoking', y='DEATH_EVENT', data=smoking_death_rate)
plt.title('Death Rate by Smoking Status')
plt.xlabel('Smoking (0: No, 1: Yes)')
plt.ylabel('Death Rate')
plt.show()
```

Are there any noticeable patterns in the distribution of death events across different age groups

```python

death_by_age_group = df.groupby('age')['DEATH_EVENT'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='DEATH_EVENT', data=df)
plt.title('Distribution of Death Events Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Death Event Rate')
plt.show()
```

Is there any significant difference in ejection fraction between patients with and without diabetes

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='diabetes', y='ejection_fraction', data=df)
plt.title('Ejection Fraction by Diabetes Status')
plt.xlabel('Diabetes (0: No, 1: Yes)')
plt.ylabel('Ejection Fraction')
plt.show()
```

How does the serum creatinine level vary between patients who survived and those who did not?

```python
# Plot serum creatinine levels by death event
plt.figure(figsize=(10, 6))
sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=df)
plt.title('Serum Creatinine Levels in Survived vs. Deceased Patients')
plt.xlabel('Death Event (0: Survived, 1: Deceased)')
plt.ylabel('Serum Creatinine')
plt.show()
```
